# Task 1: Exploring and Visualizing the Iris Dataset
**Objective:**  
Load, inspect, and visualize the Iris dataset to understand data trends, distributions, and potential outliers.

---
## Import Libraries
We will use `pandas` for data manipulation, `seaborn` and `matplotlib` for visualization.

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

## Load dataset
We use seaborn's built-in dataset

```python
df = sns.load_dataset('iris')
df.head()
```

## Basic data inspection



```python
# shape of dataset
print("Dataset shape:", df.shape)

# Column names
print("Columns:", df.columns.to_list())

# info about dataset
print("Dataset info:", df.info())

# Statistical summary
print('Summary statistics:', df.describe())
```

## Data Visualization

```python
# scatter plot: sepal length Vs sepal width
# it shows the relationship between two numeric variables
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x='sepal_length',
    y='sepal_width',
    hue='species',
    palette="Set1",
    s=70
)
plt.title('Scatter Plot: Sepal Length Vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
# Histogram: Distribution of feature
import matplotlib.pyplot as plt

df.hist(figsize = (10,7), color='lightblue', edgecolor='black')
plt.suptitle('Histogram of iris dataset feature')
plt.show()
```

```python
# Box plot: identify outliers
plt.figure(figsize=(10,6))
sns.boxplot(
    data=df,
    palette='Pastel1'
)
plt.title('Boxplot of iris dataset feature')
plt.show()
```

# Step 5: Insights
 1. Setosa has smaller petals compared to Versicolor and Virginica.
 2. Virginica has the largest sepal and petal sizes.
 3. Scatter plot shows clear separation of Setosa from others.
 4. Box plots indicate minor outliers in sepal width and petal width.
 5. Histograms show distribution shapes: mostly normal with some skew.
